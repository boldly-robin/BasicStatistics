## 2. 데이터와 표본분포
- 데이터 편향을 최소화하기 위해 표본추출의 필요성은 더욱 커지고 있다. 

### 1. 랜덤표본추출과 표본편향
1. 표본 : 더 큰 데이터 집합으로부터 얻은 데이터의 부분집합
    - 표본평균 
    
2. 모집단 : 어떤 데이터 집합을 구성하는 전체 대상 혹은 전체집합
    - 모평균  
    
3. 추출방법
    - 임의표집 : 랜덤표본추출, 무작위로 표본을 추출한 것, 모든 원소는 동일한 확률로 뽑히게 된다. 
        - 단순임의표본 : 단순랜덤표본, 모집단 층화없이 랜덤표본추출로 얻은 표본
        - 복원추출 
        - 비복원추출 
    - 층화표집 : 층화표본추출, 모집단을 층으로 나눈 뒤 각 층에서 무작위로 표본을 추출
    
4. 표본편향 : 모집단을 잘못 대표하는 표본, 모집단과 표본 사이의 차이가 유의미할만큼 크고, 첫번째 표본과 동일한 방식으로 추출된 다른 샘플들에서도 이 차이가 계속될 것으로 예상될 때, 표본편향이 발생했다고 볼 수 있다. 
    - 대표성 : 얼마나 모딥단을 잘 대표하는냐, 데이터의 품질이란 완결성, 형식의 일관성, 깨끗함, 데이터의 정확성을 의미한다.
    - 비임의 : 아무리 랜덤표본이라 해도, 어떤 표본도 모집단을 정확히 대표할 수 없다. 
    
5. 크기와 품질 : 데이터의 크기가 무조건 큰 것이 좋은 것은 아니다.
    - 실제로 연관된 레코드의 수가 수천대만 돼도 효과적일 수 있다. 
    - 데이터가 큰것이 더 좋을 경우에는 데이터가 크고, 동시에 희박할 경우이다. 구글에서 반드시 단 하나의 특정 결과를 원할 때와 같이, 데이터가 많이 축적될수록 그것이 더 좋은 결과를 반환할 수 있게 된다. 
    
    
    
### 2. 선택편향

1. 선택편향 : 데이터를 의식적으로든, 선택적으로 고르는 관행을 의미 -> 오해의 소지가 있거나 단편적인 결과를 얻게 된다.
    - 데이터스누핑 : 뭔가 흥미로운 것을 찾아 광범위하게 데이터를 살피는 것
    - 방대한 검색 효과 : 중복데이터모델링이나 너무 많은 예측변수를 고려하는 모델링에서 비롯되는 편향 혹은 비재현성
    - 이를 방지하기 위해 순열검정(목푯값 섞기)를 통해 예측들을 검증한다. 
2. 평균으로의 회귀 : 주어진 어떤변수를 연속적으로 측정했을 때 나타나는 현상, 예외적인 경우가 관찰되면 그 다음에는 중간 정도의 경우가 관찰되는 경우가 있다. 따라서, 예외를 너무 특별히 생각하고 의미를 부여하면 선택편향으로 이어질 수 도 있다. 
3. 가설을 구체적으로 명시하고, 랜덤표본추출 원칙에 따라 데이터를 수집하면 편향을 피할 수 있다. 

### 3. 통계학에서의 표본분포
1. 표본통계량 : 더 큰 모집단에서 추출된 표본데이터들로부터 얻은 측정지표, 표본통계량의 분포
2. 데이터분포 : 어떤 데이터 집합에서의 각 개별 값의 도수분포, 개별데이터포인트의 분포
3. 표본분포 : 여러 표본들 혹은 재표본들로부터 얻은 표본통계량의 도수분포 (데이터분포 != 표본분포)
4. 중심극한정리 : 표본크기가 커질수록 표본분포가 정규분포를 따르는 경향, 모집단이 정규분포가 아니더라도 표본크기가 충분하고 데이터가 정규성을 크게 이탈하지 않는 경우 여러 표본에서 추출한 평균은 종 모양의 정규곡선을 따른다. 
    - 중심극한정리덕분에 신뢰구간이나 가설검정을 계산하는 데 t분포와 같은 정규근사 공식을 사용할 수 있다. 
    - 중심 극한 정리는 어떤 분포를 따르는 확률 변수든 간에 해당 확률 변수가 복수인 경우 그 합은 정규 분포와 비슷한 분포를 이루는 현상을 말한다.
5. 표준오차 : 여러표본들로부터 얻은 표본통계량의 변량 (!= 개별데이터값들의 변량을 의미하는 표준편차) 
    - 표본크기가 커지면 표준오차가 줄어든다. 표준오차와 표본크기 사이의 관계를 n제곱근의 법칙이라고 한다. 즉, 표준오차를 2배로 줄이려면 표본 크기를 4배 증가시켜야 한다.
    - 표준오차를 측정할 때 고려해야 하는 사항
        - 모집단에서 완전히 새로운 샘플들을 많이 수집해야 한다. 
        - 각각의 새 샘플에 대한 통계량을 계산해야 한다. 
        - 2단계에서 얻은 통계량의 표준편차를 계산한다. 이것을 표준오차의 추정치로 사용한다.

### 4. 부트스트랩
- 표준오차를 추정하기 위해 새 샘플을 수집하는 방식은 불가능하다. 이 때 사용하는 것이 부트스트랩이다.
- 부트스트랩 : 현재 있는 표본에서 추가적으로 표본을 복원추출하고 각 표본에 대한 통계량과 모델을 다시 계산하는 절차이다. 이 때 반드시 데이터나 표본통계량이 정규분포를 따라야 한다는 가정은 필요하지 않다. 
- 재표본추출 과정
    - 샘플 값을 하나 뽑아서 기록하고 제자리에 놓는다
    - n번 반복한다
    - 재표본추출된 값의 평균을 기록한다
    - 1~3단계를 r번 반복한다. 
    - 이를 이용해 표준편차(표본평균의 표준오차)를 계산하고, 히스토그램을 그리고 신뢰구간을 찾는다
- 분류 및 회귀트리를 사용할 때, 여러 부트스트랩 샘플을 가지고 트리를 여러 개 만든 다음 각 트리에서 나온 예측값을 평균내는 것이 일반적으로 단일 트리를 사용하는 것보다 효과적이다. ==> 배깅


### 5. 신뢰구간
- 신뢰구간 : 같은 모집단으로부터 같은 방식으로 얻은 관심 통계량을 포함할 것으로 예상되는 신뢰구간의 백분율
- 구간끝점 : 신뢰구간의 최상위, 최하위 끝점

```python
# 부트스랩을 10,000번 반복해서 와인의 품질이 5미만인 것과 5이상인 것의 잔당 차이의 99% 신뢰구간을 구하겠다.
iterationNum = 10000
quality = []
for _ in range(iterationNum):
    bootSample = wine.sample(300, replace=True) # 복원 추출
    under5mean = bootSample[bootSample['quality'] < 5.0]['residual sugar'].mean() 
    upper5mean = bootSample[bootSample['quality'] >= 5.0]['residual sugar'].mean() 
    diff = under5mean - upper5mean
    quality.append(diff)

#  신뢰수준 99%인 평균 키 차이에 대한 신뢰구간
np.percentile(quality, 0.5), np.percentile(quality, 99.5)
```

-# 맞나?( 품질이 5미만인 와인은 품질이 5이상인 와인보다 잔당이 더 크다고 할 수 있다. )

### 6. 정규분포
- 오차 : 데이터 포인트와 예측값 혹은 평균 사이의 차이
- 표준화하다 = 정규화하다 : 평균을 빼고 표준오차로 나눈다
- Z점수 : 개별 데이터 포인트를 정규화한 값, 정규분포는 z분포라 한다. 
- 표준정규분포 : 평군 = 0, 표준편차 = 1인 정규분포, x축의 단위가 평균의 표준편차로 표현되는 정규분포를 의미한다. 
- QQ그림 : 표본분포가 정규분포에 얼마나 가까운지를 보여주는 그림
- 데이터의 99.7%는 평균의 표준편차 3배수안에 있다. 
- 데이터의 95%는 표준편차 2배수안에 있다. 
- 데이터의 68%는 표준편차 1배수안에 있다. 

```python
#qq plot
mu = 0
std = 1
rv = stats.norm(mu,std) #평균은 0, 표준편차는 1인 정규분포를 만들자

x = np.linspace(-5,5,100)
plt.plot(x,rv.pdf(x)) #pdf를 이용해 probablity density function(확률밀도함수)를 계산할 수 있다. 
```

```python
np.random.seed(0)
x = rv.rvs(100) #새로운 샘플 추출
```

```python
sns.distplot(x,kde = False, fit = stats.norm)
plt.show()
```

```python
plt.figure(figsize=(7,7))
stats.probplot(x, plot=plt)
plt.axis("equal")
plt.show()
```

- 정규분포를 따르는 경우의 QQ plot모양은 이렇듯 직선의 모습을 보인다

```python
np.random.seed(0)
x = np.random.rand(100)
plt.figure(figsize=(7,7))
stats.probplot(x, plot=plt)
plt.ylim(-0.5, 1.5)
plt.show()
```

- 정규분포를 보이지 않을 경우 이렇듯 휘어지는 모습을 보인다. 

```python
# 중심 극한 정리는 어떤 분포를 따르는 확률 변수든 간에
#해당 확률 변수가 복수인 경우 그 합은 정규 분포와 비슷한 
#분포를 이루는 현상을 말한다.

# 중심극한정리가 성립하는지 확인하는 과정

xx = np.linspace(-2, 2, 100)
plt.figure(figsize=(6,9))
for i, N in enumerate([1, 2, 10]):
    X = np.random.rand(1000, N) - 0.5
    S = X.sum(axis=1)/np.sqrt(N) #확률변수의 합, n이 증가할수록  정규분포에 수렴한다
    plt.subplot(3, 2, 2*i+1)
    sns.distplot(S, bins=10, kde=False, norm_hist=True)
    plt.xlim(-2, 2)
    plt.yticks([])
    plt.subplot(3, 2, 2*i+2)
    stats.probplot(S, plot=plt)
    
plt.tight_layout()
plt.show()
```

### 7. 긴 꼬리 분포
- 꼬리 : 적은 수의 극단값이 주로 존재하는 도수의 길고 좁은 부분
- 왜도 : 분포의 한쪽 고리가 반대쪽 다른 꼬리보다 긴 정도
- 실무에서는 긴 꼬리를 중요하게 생각하기도 한다. 정규분포로 예측하는 것조다 훨씬 더 긴꼬리 이론이 나타날 수 있다. 데이터가 정규분포를 따른다고 할 때 예상되는 것보다 훨씬 더 많은 극단값을 관찰할 가능성이 있다.: black swan theory

### 9. 이항분포

- 이항식 : 두 가지 결과를 갖는다

### 10. 푸아송분포와 그 외 관련 분포
- 푸아송분포 : 단위시간안에 어떤 사건이 몇 번 발생할지를 표현하는 이산확률분포
    - 독립성 :  어떤 단위 시간 또는 단위 공감에서 발생한 결과는 중복되지 않은 다른 시간이나 공간에서 발생한 결과와 서로 독립이다. 
    - 일정성 : 어떤 단위 시간 또는 단위 공간에서 발생한 확률은 그 시간 크기, 혹은 공간의 크기와 비례하고, 외부의 영향을 받지 않는다. 
    - 비집략성 : 매우 짧은 시간이나 매우 작은 공간에 두 개 이상의 경과가 동시에 발생할 확률은 0이다. 
    - 핵심 파라미터는 lambda이다. 이는 어떤 일정시간, 공간단위 안에서 발생한 평균 사건의 수를 의미한다. 
- 지수분포 : 연속확률분포, 사건이 독립적일 때, 일정시간동안 발생하는 사건의 횟수가 푸아송분포를 따른다면, 다음 사건이 일어날 때까지 대기 시간은 지수분포를 따른다. 확률변수 X가 증가할수록 그 발생활률은 작아지는 모습을 보인다. 
- 베이불 분포 : 지수분포를 확장한 것이다. 형상 파라미터 베타로 지정된 대로 발생률이 달라질 수 있다. 시간에 따라 변화하는 사건 발생률은 베이불 분포로 모델링할 수 있다.
